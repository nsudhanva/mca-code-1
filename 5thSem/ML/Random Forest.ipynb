```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('D:/Study/MCA/5thSem/ML/classi-titanic.csv')
df.head()
```

```python
df.describe()
```

```python
df['Sex'] = [1 if i == 'female' else 0 for i in df['Sex']]
df.head(100)
```

```python
y = df['Survived']
x = df.iloc[:,1:4]
# x = df[['Pclass', 'Sex', 'Age']]
# x.shape
# y.shape
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 40) 
print(x_train.shape)
print(x_test.shape)
```

```python
from sklearn.ensemble import BaggingClassifier, RandomForestClassifier, AdaBoostClassifier
```

```python
bagm = BaggingClassifier()
bagm.fit(x_train, y_train)
pred = bagm.predict(x_test)
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test, pred)
print(acc)
```

```python
rfm = RandomForestClassifier(n_estimators = 6)
rfm.fit(x_train, y_train)
pred = rfm.predict(x_test)
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test, pred)
print(acc * 100)
```

```python
adam = AdaBoostClassifier()
adam.fit(x_train, y_train)
pred = adam.predict(x_test)
from sklearn.metrics import accuracy_score, confusion_matrix
acc = accuracy_score(y_test, pred)
print(acc * 100)
print(confusion_matrix(y_test, pred))
```

```python
grid_param = {'criterion': ['gini', 'entropy'],
             'max_depth': [4, 5],
             'min_samples_leaf': [1, 2, 3],
             'min_samples_split': [4, 5],
             'n_estimators': [100, 150, 200],
             'n_jobs': [-1]}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(estimator = rfm, param_grid = grid_param, cv = 3, verbose = 2)
```

```python
grid.fit(x_train, y_train)
```

```python
y_pred = grid.predict(x_test)
```

```python
acc = accuracy_score(y_test, y_pred)
print(acc)
```
