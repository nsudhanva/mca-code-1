```python
# What is that you want to predict
# when y is numerical, then apply regression
# when y is categorical, then apply classification
# y = f(x) where x is 1, then it is single linear regression - one attribute
# y = f(x) where x is >1, then it is mutliple linear regression - multiple attributes
y = df['price']
print(type(y))
y = y.values
y = y.reshape(len(y), 1)
print(y)
```

```python
x = df['lotsize']
x = x.values
x = x.reshape(len(x), 1)
print(x)
```

```python
# categorical - classification --> Accuracy rate is found
# numerical - regression --> Error rate is found
x_train = x[:430]
x_test = x[431:]
y_train = y[:430]
y_test = y[431:]
regr = linear_model.LinearRegression()
regr.fit(x_train, y_train)
pred = regr.predict(x_test)
print(pred)
```

```python
rms = sqrt(mean_squared_error(y_test, pred))
print(rms)
```
